## Analysis of Central Park Area for Cycling Store Front Location.

The goal of this project is to explore where a cycling company should build a store near Central Park, NYC in which they can rent bicycles for use in the park while also raising brand awareness.



To start, I first chose 9 stations (highlighted in the Figure with stars). Four of the stations are located directly at Central Park while the others are located at stations that have bike lanes to allow individuals to safely get to the park.


Next, I examined the average daily total counts at the MTA stations across Spring, Summer (also tourist season), and Fall in 2021. 

The Figure highlights that there was little difference across the seasons. More importantly, the Figure illustrates that there are 3 MTA stations with significantly more traffic than the others. These popular locations are located a few blocks from Central Park.

These results suggest there are 3 potential locations that see a substatnital amount of pedestrians and would be ideal locations for building brand awareness.
